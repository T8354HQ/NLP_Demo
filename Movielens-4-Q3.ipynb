```python
def cosine(s1, s2):
    dot_product = np.dot(s1,s2)
    norm_s1 = np.linalg.norm(s1)
    norm_s2 = np.linalg.norm(s2)
    return dot_product / (norm_s1 * norm_s2)
```

```python
def most_similar(user, topN):
    similarities = []
    for i in range(len(train)):
        if user == i: continue
        sim = cosine(train.iloc[user],train.iloc[i])
        similarities.append([i, sim])
    similarities.sort(key=lambda x: x[1], reverse=True)
    return similarities[:topN]
```

```python
most_similar(1, 10)
```

```python
train['mean_rating'] = train.mean(axis=1)
```

# Rating Prediction
$\huge r = \overline{r_u} + \frac{\sum_{u_j\in N}sim(u,u_j)\cdot (r_{u_j,i}-\overline{r_{u_j}})}{\sum_{u_j\in N}sim(u,u_j)}$

```python
def rating_pred(user, item):
    top_10_sim_users = most_similar(1, 10)
    top_10_sim_users_id = [item[0] for item in top_10_sim_users]
    top_10_sim_users_similarities = [item[1] for item in top_10_sim_users]
    top_10_sim_users_rating_for_item = train.loc[top_10_sim_users_id][item]
    return train.iloc[user]['mean_rating']+ np.dot(top_10_sim_users_similarities, 
                  top_10_sim_users_rating_for_item - train.loc[top_10_sim_users_id]['mean_rating'])/np.sum(top_10_sim_users_similarities)

    # you can do in an advance way 
    # list(list(zip(*top_10_sim_users_id))[0]
    # list(list(zip(*top_10_sim_users_id))[1])
rating_pred(1,1600)
```
